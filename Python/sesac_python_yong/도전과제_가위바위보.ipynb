```python
import random
import matplotlib.pyplot as plt
import numpy as np


RSP_result_ls = []


def RSP_program():
    try:
        with open(r'RSP_result.txt','r') as file:
            RSP_result_file = file.read()
        RSP_result_ls = RSP_result_file.split()
        print('  .   .')
        print(' :',end=' ')
        for result in RSP_result_ls:
            print(result, end= ' ')

    except:
        print('   .')
        RSP_result_ls = []
    
    while True:
        
        RSP_chioce_ls_computer = ['','','']
        
        user_input = input()
        RSP_computer = random.choice(RSP_chioce_ls_computer)
        RSP_result = ''
        
        ##  ##
        if user_input =='':
            print(' .')
            try:
                with open('RSP_result.txt','w') as file:
                    for item in RSP_result_ls:
                        file.write(item + ' ')
                print('  .  : RSP_result.txt')
            except:
                print('  error') 
            break
        
        ##  ##
        elif user_input =='':
            win_count = RSP_result_ls.count('')
            total_count = RSP_result_ls.count('') +  RSP_result_ls.count('') +  RSP_result_ls.count('')
            win_rate = round(win_count / total_count,2) * 100 
            print(': ',win_rate,'%')
        
        ##  ##
        elif user_input == '':
            graph_x_count = np.arange(3)
            result = ['win', 'draw', 'lose']
            values = [RSP_result_ls.count(''), RSP_result_ls.count(''), RSP_result_ls.count('')]

            plt.bar(graph_x_count, values)
            plt.xticks(graph_x_count, result)

            plt.show()
        
        ##  ##
        elif user_input in RSP_chioce_ls_computer:
            if RSP_computer == '':
                if user_input == '':
                    RSP_result = ''
                elif user_input == '':
                    RSP_result = ''
                elif user_input == '':
                    RSP_result=''
                    
            elif RSP_computer == '':
                if user_input == '':
                    RSP_result=''
                elif user_input == '':
                    RSP_result=''
                elif user_input == '':
                    RSP_result=''
                    
            elif RSP_computer == '':
                if user_input == '':
                    RSP_result=''
                elif user_input == '':
                    RSP_result=''
                elif user_input == '':
                    RSP_result=''
            print(f'Player : {user_input}, Computer : {RSP_computer}, {RSP_result}')
            RSP_result_ls.append(RSP_result)
    return None
```

```python
if __name__ == "__main__":
    RSP_program()
```
