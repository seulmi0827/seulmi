```python
import random
import matplotlib.pyplot as plt
import numpy as np


RSP_result_ls = []


def RSP_program():
    try:
        with open(r'RSP_result.txt','r') as file:
            RSP_result_file = file.read()
        RSP_result_ls = RSP_result_file.split()
        print('이전 실행결과가 존재합니다. 이전 결과를 불러옵니다.')
        print('이전 결과:',end=' ')
        for result in RSP_result_ls:
            print(result, end= ' ')

    except:
        print('이전 실행결과가 존재하지 않습니다.')
        RSP_result_ls = []
    
    while True:
        
        RSP_chioce_ls_computer = ['가위','바위','보']
        
        user_input = input()
        RSP_computer = random.choice(RSP_chioce_ls_computer)
        RSP_result = ''
        
        ## 종료 ##
        if user_input =='종료':
            print('프로그램을 종료합니다.')
            try:
                with open('RSP_result.txt','w') as file:
                    for item in RSP_result_ls:
                        file.write(item + ' ')
                print('결과 파일이 저장되었습니다. 파일명 : RSP_result.txt')
            except:
                print('파일 저장 error') 
            break
        
        ## 승률 ##
        elif user_input =='승률':
            win_count = RSP_result_ls.count('승리')
            total_count = RSP_result_ls.count('승리') +  RSP_result_ls.count('패배') +  RSP_result_ls.count('무승부')
            win_rate = round(win_count / total_count,2) * 100 
            print('현재승률: ',win_rate,'%')
        
        ## 그래프 ##
        elif user_input == '그래프':
            graph_x_count = np.arange(3)
            result = ['win', 'draw', 'lose']
            values = [RSP_result_ls.count('승리'), RSP_result_ls.count('무승부'), RSP_result_ls.count('패배')]

            plt.bar(graph_x_count, values)
            plt.xticks(graph_x_count, result)

            plt.show()
        
        ## 가위바위보 ##
        elif user_input in RSP_chioce_ls_computer:
            if RSP_computer == '가위':
                if user_input == '가위':
                    RSP_result = '무승부'
                elif user_input == '바위':
                    RSP_result = '승리'
                elif user_input == '보':
                    RSP_result='패배'
                    
            elif RSP_computer == '바위':
                if user_input == '가위':
                    RSP_result='패배'
                elif user_input == '바위':
                    RSP_result='무승부'
                elif user_input == '보':
                    RSP_result='승리'
                    
            elif RSP_computer == '보':
                if user_input == '가위':
                    RSP_result='승리'
                elif user_input == '바위':
                    RSP_result='패배'
                elif user_input == '보':
                    RSP_result='무승부'
            print(f'Player : {user_input}, Computer : {RSP_computer}, {RSP_result}')
            RSP_result_ls.append(RSP_result)
    return None
```

```python
if __name__ == "__main__":
    RSP_program()
```
