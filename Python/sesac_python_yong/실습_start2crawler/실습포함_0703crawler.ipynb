```python
from bs4 import BeautifulSoup

res = requests.get(r'https://youth.seoul.go.kr/infoData/sprtInfo/list.do?sprtInfoId=&key=2309130006&pageIndex=1&orderBy=regYmd+desc&recordCountPerPage=8&sc_ctgry=PDS_12_YC&sw=&sc_rgn=%EC%84%9C%EB%B6%81&sc_sarea=%EC%9D%80%ED%8F%89%EA%B5%AC&sc_trgtNm=TN_000&viewType=on&sc_aplyPrdEndYmdUseYn=&cntrLa=37.566761128870546&cntrLo=126.97862963872868&neLat=37.566761128870546&neLng=126.97862963872868&swLat=37.566761128870546&swLng=126.97862963872868&mapLvl=6&sarea=#none')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
# <div class="name"> </div>
soup.select('div')
soup.select('.name')

soup.select('div.name')

soup.select('div .name') #div 태그 안에 있는 class가 name인 태그
soup.select('div > .name')

# <a href="https://naver.com" id="naver-link" class="third-party-link"> </a>
soup.select_one('a')
soup.select_one('#naver-link')
soup.select_one('a#naver-link')

# 속성으로 가져오기
soup.select('[속석명 = "값"]')
soup.select('[속성명^ = "값"]') # 값으로 시작하는 애들 가져오기
soup.select('[속성명$ = "값"]') # 값으로 끝나는 애들 가져오기
soup.select('[속석명* = "값"]') # 값을 포함하는 애들 가져오기
```

```python
soup = BeautifulSoup(res.text,'html.parser')
```

```python
feeds = soup.select('div.feed-item')
feeds
```

```python

for feed in feeds:
    temp = dict()
    
    temp['click'] = feed.select_one('a.item-overlay').attrs['onclick']
    temp['state'] = feed.select_one('span.state').text
    temp['caegory'] = feed.select_one('.cate').text.strip()
    temp['name'] = feed.select_one('.name').text.strip()
    
    result.append(temp)

result
```

```python
key = open('./secret','r').read()
key
```

```python
url = f'http://apis.data.go.kr/1360000/VilageFcstInfoService_2.0/getUltraSrtNcst?serviceKey={key}&numOfRows=10&pageNo=1&base_date=20250703&base_time=0600&nx=55&ny=127&dataType=JSON'
    
res = requests.get(url)
res
```

```python
res.text
```

```python
data = json.loads(res.text)
result = {}
key_ls = []
value_ls = []
for i in data['response']['body']['items']['item']:
    key_ls.append(i['category'])
    value_ls.append(i['obsrValue'])
    result[i['category']] = i['obsrValue']

print(len(key_ls))
print(len(value_ls))
print(result)
```
