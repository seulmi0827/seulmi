## 원래가격 역연산

```python
print(response.text.find('와플'))
data = response.json()
result_ls = []

for page in range(1,52):
    my_page = data['data']['modules'][page]['items']
    for i in range(len(my_page)):
        temp_dict = {}
        if 'brandName' in my_page[i]['info'].keys():
            my_tuple = my_page[i]['info']
            temp_dict['brandName'] = my_tuple['brandName']
            temp_dict['productName'] = my_tuple['productName']
            temp_dict['finalPrice'] = my_tuple['finalPrice']
            temp_dict['discountRatio'] = my_tuple['discountRatio']
            temp_dict['originprice'] = round(temp_dict['finalPrice']/((100 - my_tuple['discountRatio'])/100))
            result_ls.append(temp_dict)
            
len(result_ls)
```

## 실습 네이버웹툰

```python
response = requests.get('https://comic.naver.com/api/webtoon/titlelist/weekday?order=user')
data = response.json()
result_ls = []
url = 'https://comic.naver.com/webtoon/list?titleId='
for key in data['titleListMap'].keys():
    temp_dict = {}
    for i in range(len(data['titleListMap'][key])):
        temp_dict['제목'] = data['titleListMap'][key][i]['titleName']
        temp_dict['링크'] = url + str(data['titleListMap'][key][i]['titleId'])
        if key == 'MONDAY':
            weekday_kor = '월'
        elif key == 'TUESDAY':
            weekday_kor = '화'
        elif key == 'WEDNESDAY':
            weekday_kor = '수'
        elif key == 'THURSDAY':
            weekday_kor = '목'
        elif key == 'FRIDAY':
            weekday_kor = '금'
        elif key == 'SATURDAY':
            weekday_kor = '토'
        elif key == 'SUNDAY':
            weekday_kor = '일'                    
        temp_dict['요일'] = weekday_kor
        print(temp_dict)
        
```
